```python
import pandas as pd
import numpy as np
import altair as alt
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('exoplanet.csv')
df
```

```python
count_agg = df['disc_year'].value_counts().reset_index()

sns.lineplot(data = count_agg, x = 'disc_year', y = 'count')
plt.xlabel('Discovery Year')
plt.ylabel('Number of New Exoplanet Found')
plt.title('Discovery of New Exoplanet Throughout the Year')
```

```python
alt.data_transformers.disable_max_rows()
df.info()
```

```python
mean_agg = df.groupby('disc_year')['sy_dist'].mean().reset_index()

sns.lineplot(data = mean_agg, x = 'disc_year', y = 'sy_dist')
plt.xlabel('Discovery Year')
plt.ylabel('Average Distance From Earth To Exoplanet (in parsecs)')
plt.title('Expanding Discovery Distance of Exoplanet Throughout the Year')
```

```python
sns.countplot(data = df, y = 'discoverymethod')
plt.xlabel('Frequency')
plt.ylabel('Discovery Method')
plt.title('What Kind of Method Are Used To Discover New Exoplanet?')
```
